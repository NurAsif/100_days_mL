```python
def get_batsman_runs(batsman_name):
    batsman = data[data['batsman'] == batsman_name]  # Boolean mask for rows where batsman matches the given name
    return batsman.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)  # Sum of runs scored by the batsman for each team
```

```python
over = data ['over']>15
death_over = data[over]  # Filtered DataFrame for overs greater than 15
```

```python
death_over
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()  # Count of runs scored by each batsman
```

```python
all_batsman
```

```python
y = all_batsman[all_batsman > 200]  # Filter batsmen who have faced more than 200 balls in death overs
```

```python
y
```

```python
x = all_batsman > 200 # Filter batsmen who have faced more than 200 balls in death overs
all_batsman[x]  # Filtered DataFrame for batsmen who have faced more than 200 balls in death overs
```

```python
all_batsman[x].index  # Index of batsmen who have faced more than 200 balls in death overs
```

```python
all_batsman[x].values  # Values of batsmen who have faced more than 200 balls in death overs
```

```python
batsman_list = all_batsman[x].index.tolist()  # Convert index of batsmen who have faced more than 200 balls in death overs to a list
```

```python
# Runs scored by all these 43 batsmen in death overs
# Balls faced by these 43 batsmen in death overs
data['batsman'].isin(batsman_list)  # Boolean mask for rows where batsman is in the list of batsmen who have faced more than 200 balls in death overs
```

```python
data[data['batsman'].isin(batsman_list) ]  # Filtered DataFrame for rows where batsman is in the list of batsmen who have faced more than 200 balls in death overs
```

```python
data[data['batsman'].isin(batsman_list) ].shape
```

```python
final = data[data['batsman'].isin(batsman_list) ]  # Final DataFrame for rows where batsman is in the list of batsmen who have faced more than 200 balls in death overs
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum() # Sum of runs scored by each batsman in death overs
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()  # Count of balls faced by each batsman in death overs
```

```python
sr = (runs / balls) * 100  # Calculate strike rate for each batsman in death overs
```
